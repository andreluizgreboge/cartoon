```python
# carregar bibliotecas
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
carregar imagem
img = cv2.imread('fjpg.jpg')
cv2_imshow(img)
```

```python
# aqui vamos deixar os "contornos" mais marcados
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_blur = cv2.medianBlur(gray, 3)
edges = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                cv2.THRESH_BINARY,
                                9, 3)
cv2_imshow(edges)

```

```python
# aqui aletramos a cor da foto, "misturando" os pixels.
cor = cv2.bilateralFilter(img, d=9, sigmaColor=9900
                          ,sigmaSpace=9900)
```

```python
cartoon = cv2.bitwise_and(cor, cor, mask=edges)
cv2_imshow(cartoon)
```
